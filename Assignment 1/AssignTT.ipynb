```python
gender_counts = df['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Pclass'].value_counts().sort_index().plot.pie(
    autopct='%1.1f%%',
    labels=['Class 1', 'Class 2', 'Class 3'],
    startangle=90
)

plt.title("Passenger Class Distribution")
plt.ylabel('')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')

# Bar chart of 'Embarked' (Port of Embarkation)
sns.countplot(x='Embarked', data=df, palette='pastel')
plt.title('Number of Passengers by Port of Embarkation')
plt.xlabel('Port (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare (Dot Graph)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot of Age vs Fare, colored by survival
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='Set1')

plt.title('Titanic: Age vs Fare (Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
